```python
import pickle

def open_file(file):
  with open(file, 'rb') as x:
    return pickle.load(x)

X_train_filtered= open_file('X_train_filtered.pkl')
y_train1 = open_file('y_train1.pkl')
```

```python
import pymc3 as pm
```

```python
with pm.Model() as logistic_model:
    # Priors
    tau = pm.HalfCauchy('tau', beta=1)
    lambda_ = pm.HalfCauchy('lambda_', beta=1, shape=1092)
    beta = pm.Normal('beta', mu=0, sd=lambda_ * tau, shape=1092)

    # Linear combination
    linear_combination = pm.math.dot(X_train_filtered, beta)

    # Logistic transformation
    p = pm.math.sigmoid(linear_combination)

    # Likelihood
    observed = pm.Bernoulli('observed', p=p, observed=y_train1)

    # Sampling
    trace = pm.sample(300, tune=250, chains=2, cores=1, target_accept=0.9)

# Print summary statistics of the posterior distribution
print(pm.summary(trace))

# Plot the trace of the coefficients
pm.traceplot(trace)
```
